# 3- Visualizing Data With Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
# Dummy data for S&P 500 and FTSE 100 indices
dates = ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01']
sp_500 = [2800, 2850, 2900, 2950, 3000]  # Replace with actual S&P 500 data
ftse_100 = [7200, 7250, 7300, 7350, 7400]  # Replace with actual FTSE 100 data

# Plotting on separate graphs
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)  # S&P 500
plt.plot(dates, sp_500, color='blue')
plt.title('S&P 500 Index')
plt.xlabel('Date')
plt.ylabel('Value')
```

```python
plt.subplot(2, 1, 2)  # FTSE 100
plt.plot(dates, ftse_100, color='red')
plt.title('FTSE 100 Index')
plt.xlabel('Date')
plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
# Plotting on a single graph as multiple line chart
plt.figure(figsize=(8, 5))
plt.plot(dates, sp_500,  color='blue', label='S&P 500')
plt.plot(dates, ftse_100, color='red', label='FTSE 100')
plt.title('S&P 500 vs FTSE 100 Indices')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# For scatter plot and bar chart with dummy data
import numpy as np
```

```python
# Generating random data for scatter plot and bar chart
np.random.seed(0)
x = np.random.randn(40)
y = np.random.randn(40) * 2
heights = np.random.randint(1, 10, size=40)

# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='green')
plt.title('Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(np.arange(len(heights)), heights)
plt.title('Random Bar Chart')
plt.xlabel('Bars')
plt.ylabel('Heights')
plt.show()
```
